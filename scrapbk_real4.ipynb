```python
# -------------------------------
# Step 1: Install dependencies
# -------------------------------
!pip install --quiet requests beautifulsoup4 pandas

# -------------------------------
# Step 2: Import libraries
# -------------------------------
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# -------------------------------
# Step 3: Set up URLs and storage
# -------------------------------
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
all_books = []

# -------------------------------
# Step 4: Scrape all 50 pages
# -------------------------------
for page in range(1, 51):  # 50 pages total
    url = base_url.format(page)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    books = soup.select(".product_pod")
    print(f"Page {page}: {len(books)} books found")

    for book in books:
        title = book.h3.a["title"]
        price = book.select_one(".price_color").text
        availability = book.select_one(".availability").text.strip()
        rating = book.select_one(".star-rating")["class"][1]  # e.g., "Three"

        # -------------------------------
        # Step 4a: Open detail page for category
        # -------------------------------
        detail_url = "https://books.toscrape.com/catalogue/" + book.h3.a["href"]
        detail_r = requests.get(detail_url)
        detail_soup = BeautifulSoup(detail_r.text, "html.parser")
        category = detail_soup.select("ul.breadcrumb li a")[-1].text

        all_books.append([title, price, availability, rating, category])
        time.sleep(0.1)  # small delay to be polite to server

# -------------------------------
# Step 5: Save to CSV
# -------------------------------
df = pd.DataFrame(all_books, columns=["Title", "Price", "Availability", "Rating", "Category"])
df.to_csv("books_full.csv", index=False)
print("Scraping complete. Total books:", len(all_books))

# -------------------------------
# Step 6: Download CSV
# -------------------------------
from google.colab import files
files.download("books_full.csv")
```

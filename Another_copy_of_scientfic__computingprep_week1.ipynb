```python
import numpy as np

# Define a vector
v = np.array([1, 2, 3])

# Define a matrix
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

print("Vector v:", v)
print("Matrix A:\n", A)
```

```python
B = np.array([[1, 2, 3],
              [4, 5, 6]])

result = A @ B   # matrix multiplication

print("A shape:", A.shape)
print("B shape:", B.shape)
print("A @ B =\n", result)
```

```python
A = np.array([[3, 1],
              [1, 2]])

b = np.array([9, 8])

x = np.linalg.solve(A, b)

print("Solution x:", x)
```
